```python
algoritmo "desconto_progressivo"
var
   valor, desconto, final: real
   pagamento: caractere
inicio
   escreva("Valor do produto: ")
   leia(valor)
   escreva("Forma de pagamento (dinheiro/cartao): ")
   leia(pagamento)

   se valor < 100 entao
      desconto <- 0.05
   senao
      se valor <= 500 entao
         desconto <- 0.1
      senao
         desconto <- 0.15
      fimse
   fimse

   se pagamento = "dinheiro" entao
      desconto <- desconto + 0.05
   fimse

   final <- valor * (1 - desconto)
   escreva("Valor final: R$ ", final:4:2)
fimalgoritmo
```

```python
algoritmo "fibonacci"
var
   n, a, b, temp, i: inteiro
inicio
   leia(n)
   a <- 0
   b <- 1
   para i de 1 ate n faca
      escreva(a, " ")
      temp <- a + b
      a <- b
      b <- temp
   fimpara
fimalgoritmo
```

```python
algoritmo "fatorial"
var
   n, i, fat: inteiro
inicio
   leia(n)
   fat <- 1
   para i de 1 ate n faca
      fat <- fat * i
   fimpara
   escreva("Fatorial (FOR): ", fat)

   fat <- 1
   i <- 1
   enquanto i <= n faca
      fat <- fat * i
      i <- i + 1
   fimenquanto
   escreva("Fatorial (WHILE): ", fat)
fimalgoritmo
```

```python
algoritmo "contagem"
var
   n, positivos, negativos, nulos: inteiro
inicio
   positivos <- 0
   negativos <- 0
   nulos <- 0
   repita
      leia(n)
      se n = 0 entao
         saia
      fimse
      se n > 0 entao
         positivos <- positivos + 1
      senao
         se n < 0 entao
            negativos <- negativos + 1
         senao
            nulos <- nulos + 1
         fimse
      fimse
   ate falso
   escreva("Positivos: ", positivos, " Negativos: ", negativos, " Nulos: ", nulos)
fimalgoritmo
```

```python
algoritmo "menu_areas"
var
   opcao: inteiro
   base, altura, lado, raio, area: real
inicio
   escreva("1 - Quadrado\n2 - Triângulo\n3 - Círculo\n4 - Retângulo\n")
   leia(opcao)

   escolha opcao
      caso 1
         escreva("Digite o lado: ")
         leia(lado)
         area <- lado * lado
         escreva("Área do quadrado: ", area)
      caso 2
         escreva("Digite a base e a altura: ")
         leia(base, altura)
         area <- (base * altura) / 2
         escreva("Área do triângulo: ", area)
      caso 3
         escreva("Digite o raio: ")
         leia(raio)
         area <- 3.14159 * (raio * raio)
         escreva("Área do círculo: ", area)
      caso 4
         escreva("Digite a base e a altura: ")
         leia(base, altura)
         area <- base * altura
         escreva("Área do retângulo: ", area)
      outrocaso
         escreva("Opção inválida!")
   fimescolha
fimalgoritmo
```

```python
algoritmo "medias_pares_impares"
var
   n, pares, impares, soma_pares, soma_impares, maior, menor: inteiro
inicio
   pares <- 0
   impares <- 0
   soma_pares <- 0
   soma_impares <- 0 
   maior <- -9999
   menor <- 9999

   repita
      escreva("Digite um número (negativo para sair): ")
      leia(n)

      se n < 0 entao
         saia
      fimse

      se n > maior entao
         maior <- n
      fimse
      se n < menor entao
         menor <- n
      fimse

      se n % 2 = 0 entao
         pares <- pares + 1
         soma_pares <- soma_pares + n
      senao
         impares <- impares + 1
         soma_impares <- soma_impares + n
      fimse
   ate falso

   se pares > 0 entao
      escreva("Média dos pares: ", soma_pares / pares)
   fimse
   se impares > 0 entao
      escreva("Média dos ímpares: ", soma_impares / impares)
   fimse

   escreva("Maior número: ", maior)
   escreva("Menor número: ", menor)
fimalgoritmo
```

```python
algoritmo "sistema_votacao"
var
   voto: caractere
   votosA, votosB, votosC, votosN: inteiro
inicio
   votosA <- 0
   votosB <- 0
   votosC <- 0
   votosN <- 0

   repita
      escreva("Vote (A/B/C ou 0 para encerrar): ")
      leia(voto)

      se voto = "0" entao
         saia
      senao
         se voto = "A" entao
            votosA <- votosA + 1
         senao
            se voto = "B" entao
               votosB <- votosB + 1
            senao
               se voto = "C" entao
                  votosC <- votosC + 1
               senao
                  votosN <- votosN + 1
               fimse
            fimse
         fimse
      fimse
   ate falso

   escreva("Votos A: ", votosA)
   escreva("Votos B: ", votosB)
   escreva("Votos C: ", votosC)
   escreva("Votos Nulos: ", votosN)

   se votosA > votosB e votosA > votosC entao
      escreva("Vencedor: A")
   senao
      se votosB > votosC entao
         escreva("Vencedor: B")
      senao
         escreva("Vencedor: C")
      fimse
   fimse
fimalgoritmo
```

```python
algoritmo "ordenacao_manual"
var
   numeros: vetor[1..10] de inteiro
   i, j, temp: inteiro
inicio
   para i de 1 ate 10 faca
      escreva("Número ", i, ": ")
      leia(numeros[i])
   fimpara

   para i de 1 ate 9 faca
      para j de i+1 ate 10 faca
         se numeros[i] > numeros[j] entao
            temp <- numeros[i]
            numeros[i] <- numeros[j]
            numeros[j] <- temp
         fimse
      fimpara
   fimpara

   escreva("Ordem crescente: ")
   para i de 1 ate 10 faca
      escreva(numeros[i], " ")
   fimpara
fimalgoritmo
```

```python
algoritmo "funcao_mdc"
funcao inteiro mdc(a, b: inteiro)
inicio
   enquanto b <> 0 faca
      resto <- a % b
      a <- b
      b <- resto
   fimenquanto
   retorne a
fimfuncao

var
   x, y: inteiro
inicio
   escreva("Digite dois números: ")
   leia(x, y)
   escreva("MDC = ", mdc(x, y))
fimalgoritmo
```

```python
algoritmo "fatorial_recursivo"
funcao inteiro fatorial(n: inteiro)
inicio
   se n <= 1 entao
      retorne 1
   senao
      retorne n * fatorial(n - 1)
   fimse
fimfuncao

var
   num: inteiro
inicio
   escreva("Digite um número: ")
   leia(num)
   escreva("Fatorial = ", fatorial(num))
fimalgoritmo
```
