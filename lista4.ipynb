```python
algoritmo "converter_horas"
var
   horas, minutos, segundos: inteiro
inicio
   escreva("Digite o valor em horas: ")
   leia(horas)
   minutos <- horas * 60
   segundos <- horas * 3600
   escreva("Minutos: ", minutos, " Segundos: ", segundos)
fimalgoritmo
```

```python
algoritmo "reservatorio"
var
   volume, vazao, tempo: real
inicio
   escreva("Volume (litros): ")
   leia(volume)
   escreva("Vazão (litros/minuto): ")
   leia(vazao)
   tempo <- volume / vazao
   escreva("Tempo necessário: ", tempo, " minutos")
fimalgoritmo
```

```python
algoritmo "autenticacao"
var
   senha, tentativa: caractere
   i: inteiro
inicio
   senha <- "1234"
   para i de 1 ate 3 faca
      escreva("Digite a senha: ")
      leia(tentativa)
      se tentativa = senha entao
         escreva("Acesso permitido")
         saia
      fimse
   fimpara
   se tentativa <> senha entao
      escreva("Acesso bloqueado")
   fimse
fimalgoritmo
```

```python
soma = count = 0
while True:
    idade = int(input("Idade (0 para sair): "))
    if idade == 0:
        break
    soma += idade
    count += 1

print("Média das idades:", soma / count if count > 0 else 0)
```

```python
positivos = negativos = zeros = 0

while True:
    n = int(input("Número (999 para sair): "))
    if n == 999:
        break
    if n > 0:
        positivos += 1
    elif n < 0:
        negativos += 1
    else:
        zeros += 1

print("Positivos:", positivos)
print("Negativos:", negativos)
print("Zeros:", zeros)
```

```python
medias = []
for i in range(5):
    notas = [float(input(f"Nota {j+1} do aluno {i+1}: ")) for j in range(4)]
    media = sum(notas) / 4
    medias.append(media)
    print(f"Média do aluno {i+1}: {media:.2f}")

print("Média geral da turma:", sum(medias) / len(medias))
```

```python
def fatorial(n):
    fat = 1
    for i in range(1, n+1):
        fat *= i
    return fat

print(fatorial(int(input("Digite um número: "))))
```

```python
def mdc(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print("MDC:", mdc(int(input("A: ")), int(input("B: "))))
```

```python
def analisa(notas):
    return max(notas), min(notas), sum(notas) / len(notas)

notas = [float(input(f"Nota {i+1}: ")) for i in range(5)]
maior, menor, media = analisa(notas)
print(f"Maior: {maior}, Menor: {menor}, Média: {media:.2f}")
```

```python
def eh_perfeito(n):
    soma = sum(i for i in range(1, n) if n % i == 0)
    return soma == n

num = int(input("Digite um número: "))
if eh_perfeito(num):
    print("Número perfeito")
else:
    print("Não é perfeito")
```

```python
def lerVetor():
    return [float(input(f"Valor {i+1}: ")) for i in range(5)]

def maiorElemento(v):
    return max(v)

def mediaVetor(v):
    return sum(v) / len(v)

def exibirResultados(v):
    print("Maior elemento:", maiorElemento(v))
    print("Média do vetor:", mediaVetor(v))

vetor = lerVetor()
exibirResultados(vetor)
```
